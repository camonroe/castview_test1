```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
a = 0.1  # Adjust this value to control the shape of the spiral
velocity = 0.1  # Constant velocity

# Initialize time and initial position
t = 0
x = []
y = []

# Number of time steps and maximum time
num_steps = 1000
max_time = 10 * np.pi  # Adjust this value to control the length of the spiral

# Generate the spiral points
for _ in range(num_steps):
    x.append(a * t * np.cos(t))
    y.append(a * t * np.sin(t))

    # Increment time based on velocity
    t += velocity

# Plot the spiral
plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.title('Archimedean Spiral')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.grid(True)
plt.show()
```
